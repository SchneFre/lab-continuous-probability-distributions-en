## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# your code here

def uniform_rand_numbers(bottom, ceiling, count):
    return np.random.uniform(bottom, ceiling, size=count)

```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# your code here

def normally_distributed_rand(mean, standard_deviation, count):
    vals = norm.rvs(mean, standard_deviation, count)
    return vals

```

```python
# your code here

set_1 = normally_distributed_rand(10,1,1000)
set_2 = normally_distributed_rand(10,50,1000)
```

```python
# your code here
plt.hist(set_1, bins=20, edgecolor='black')  # 20 bins, black edges
plt.title("Histogram of normally Distributed Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# your code here
plt.hist(set_2, bins=20, edgecolor='black')  # 20 bins, black edges
plt.title("Histogram of normally Distributed Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

How are the two distributions different?

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here

def exp_distributed(count):
    mean = 10
    return np.random.exponential(scale=mean, size=count)

set_10  = exp_distributed(100)
set_100 = exp_distributed(1000)
set_100
```

```python
plt.hist(set_10, bins=30, edgecolor='black')
plt.title("set_10")
plt.show()
```

```python
plt.hist(set_100, bins=30, edgecolor='black')
plt.title("set_100")
plt.show()
```

How are the two distributions different?

```python
"""
small sample = noisier, large sample = smoother exponential decay
"""
```

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
from scipy.stats import expon

lambda_param = 1/10
prob_less_than_15 = expon.cdf(15, scale=1/lambda_param)
prob_less_than_15
```

```python
# your answer here
# Hint: This is same as saying P(x<15)
# P(x<15) = 77,7%
```

What is the probability that the customer will spend more than 15 minutes

```python
# your code here
prob_more_than_15 = 1 - prob_less_than_15
prob_more_than_15
```
